```python
from keras.models import load_model
import os
import subprocess
```

```python
directory = 'newModel'

for fileName in os.listdir(directory):
    modelPath = os.path.join(directory, fileName)
    print(modelPath)
    model = load_model(modelPath)
    tfjs_target_dir = 'newModelJs/'+fileName.replace('.h5', '')
    command = []
    command.append("tensorflowjs_converter")
    command.append("--input_format=keras")
    command.append(modelPath)
    command.append(tfjs_target_dir)
    subprocess.call(command)
```
